```python
import pandas as pd
import numpy as np
import textwrap
```

### 1.1

```python
emails = pd.read_csv("../Data/all-data/lingspam-emails.csv.bz2",sep='\t')
emails.head()
```

### 1.2

```python
emails = emails.dropna(subset=['spam', 'message'])
emails.head()
```

### 2.1

```python
list_of_words = ['viagra', 'deadline', 'million', 'and', 'drink', 'dear', 'update', 'billion', 'food', 'pizza']
```

### 2.2

```python
for w in list_of_words:
    emails[w] = emails.message.str.lower().str.contains(w)
```

```python
X = emails.iloc[:, 3:]
Y = emails.spam
X
```

### 2.3

```python
from sklearn.model_selection import train_test_split

Xt, Xv, Yt, Yv = train_test_split(X, Y)
```

```python
Yv
```

### 3.1

- `P_S`: Pr(category = S)
- `P_NS`: Pr(category = NS)
- `P_W`: Pr(word = 1)
- `P_NW`: Pr(word = 0)
- `P_W_S`: Pr(word = 1 | category = S)
- `P_NW_S`: Pr(word = 0 | category = S)
- `P_W_NS`: Pr(word = 1 | category = NS)
- `P_NW_NS`: Pr(word = 0 | category = NS)

### 3.2

```python
P_S = np.mean(Yt)
P_NS = 1 - P_S
print("Probability of spam: ", P_S)
print("Probability of non-spam: ", P_NS)
```

### 3.3

```python
for w in list_of_words:
    print(f'{w}: P_W = {round(np.mean(Xt[w]), 3)}')
    print(f'{w}: P_NW = {round(np.mean(Xt[w] == False), 3)}')
```

### 3.4

```python
for w in list_of_words:
    print(f'{w}: P_W_S = {round(np.mean(Xt[w][Yt]), 3)}')
    print(f'{w}: P_W_NS = {round(np.mean(Xt[w][Yt == False]), 3)}')
```

### 3.5

```python
for w in list_of_words:
    print(f'{w}: P_S_W = {np.mean(Xt[w][Yt]) * P_S / (np.mean(Xt[w][Yt]) * P_S + np.mean(Xt[w][Yt == False] * P_NS))}')
    print(f'{w}: P_S_NW = {np.mean((Xt[w] == False)[Yt]) * P_S / (np.mean((Xt[w] == False)[Yt]) * P_S + np.mean((Xt[w] == False)[Yt == False]) * P_NS)}')
```

### 3.6

Only `P_W` `P_NW` `P_S` `P_NS` have to sum to unity. Others are not. This is because no situations when both probabilities happen at the same time. The P_W_S cannot be added up to 1 with P_W_NS, since thre are no extra cases outside of these probabilities, happening alongside them.

### 3.7 & 3.8

```python
from sklearn.metrics import confusion_matrix

for w in list_of_words:
    P_S_W = np.mean(Xt[w][Yt]) * P_S / (np.mean(Xt[w][Yt]) * P_S + np.mean(Xt[w][Yt == False] * P_NS))
    P_S_NW = np.mean((Xt[w] == False)[Yt]) * P_S / (np.mean((Xt[w] == False)[Yt]) * P_S + np.mean((Xt[w] == False)[Yt == False]) * P_NS)
    
    P = np.where(Xv[w], P_S_W, P_S_NW)
    y_hat = P > 0.5
    
    tn, fp, fn, tp = confusion_matrix(Yv, y_hat).ravel()
    precision = tp / (tp + fp)
    recall = tp / (tp + fn)
    
    print("Word: ", w)
    print(P_S_W, P_S_NW)
    print("Confusion Matrix:\n", confusion_matrix(Yv, y_hat))
    print("Accuracy: ", np.mean(y_hat == Yv))
    print("Precision: ", precision)
    print("Recall: ", recall, "\n")
```

### 3.9

Most of the emails are identified as negative(spam emails) and in some extreme cases that no emails are identified as spam. Thus, recall and making the precision are extremely low. The accuracy looks high when the emails in the files mostly are classfied as non-spam.

### 3.10

For the overall accuracy of the model is high, which is over 80%. However, except from millions and billions, in either cases, both of the probability are under 0.5 making all emails will be classified as negative(spam), which means that it is not a good model to identify positive examples

The reason of low recall while other indicators do not look back is because the model only correctly predicts few of actual spam emails as the spam. The accuracy is high due to the reason that the predictions of ratio of the spam and non spam emails are high compared with all predictions.
